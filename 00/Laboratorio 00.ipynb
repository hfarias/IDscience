En la carpeta `datascience`ejecutamos el comando de :

Que de salir todo correctamente nos entrega el siguiente mensaje:

>A `Vagrantfile` has been placed in this directory. You are now
ready to `vagrant up` your first virtual environment! Please read
the comments in the Vagrantfile as well as documentation on
`vagrantup.com` for more information on using Vagrant.


Ahora debemos configurar el archivo `Vagrantfile` para :

>  sudo vim Vagrantfile

Agregarmos la siguiente linea de configuración al archivo:

> config.vm.network :forwarded_port, host: 8888, guest: 8888

Lo primero es levantar la maquina con el siguiente comando:

> **vagrant up**

Ahora nos conectamos a nuestro ubuntu, con el siguiente comando:

> **vagrant ssh**

Para finalizar esta etapa algunos comandos utiles de vagrant:
> **vagrant suspend** - Suspende la máquina de modo que no se utilizarán más recursos.

> **vagrant resume** -  Restaurado el estado de la maquina, post vagrant suspend

> **vagrant halt** -  Apaga completamente la máquina.

> **vagrant reload** -  Vuelve a cargar la configuracion de la maquina si cambiamos la configuracion del Vagrantfile.

## Nivel 2: Virtualenv
___
 Ian Bicking, creador de Virtualenv:

> Virtualenv is a Python tool used to create isolated environments for Python in which you can install packages without interfering with the other virtualenvs nor with the system Python's packages.


[Documentación oficial](http://virtualenv.readthedocs.org/en/latest/)

El primer paso es instalar **pip**, esto se hace con el siguiente comando (en la maquina vagrant):

> sudo apt-get install python-pip

Vemos que paquetes tiene originalmente instalado el SO :

> pip freeze > paquetes.txt

Ahora se procede a instalar **virtualenv**: 
> sudo pip install virtualenv

Una vez instalado **virtualenv**, se procede a crear en el entorno para el analisis de datos:

> virtualenv datascience

Ahora procedemos a activar el entorno:

> source datascience/bin/activate

Con lo que el promt deberia cambiar a :

> vagrant@precise32:~$ source datascience/bin/activate

> **(datascience)**vagrant@precise32:~$ 


vemos que paquetes tiene originalmente instalado el SO :

> pip freeze > paquetes_entorno.txt

Estos son:

- argparse==1.2.1
- wsgiref==0.1.2

Finalmente para salir del entorno se ingresa el comando `deactivate`, **es importante destacar que todo lo que se instale mientras el entorno de virtualenv este activo solo esta estara instalado en el entorno.**
